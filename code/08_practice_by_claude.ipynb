— if/elif/else logic

Write a program that takes a student's score (0-100) as input and prints their letter grade:

90+ → "A"
80-89 → "B"
70-79 → "C"
60-69 → "D"
below 60 → "F"

Test it with at least 3 different scores.

```python
x=int(input("Enter the score of student in 100: "))
if(x>=90):
    print("A")
elif (x>=80):
    print("B")
elif (x>=70):
    print("C")
elif (x>=60):
    print("D")
else:
    print("F")
```

Q4 

— match-case (the new thing you just learned)

Write a function day_type(day) that takes a day name as a string and returns:

"Weekday" for Monday through Friday

"Weekend" for Saturday or Sunday

"Invalid day" for anything else



```python
def day_type(day):
    match day:
        case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":
            return "Weekday"
        case "Saturday" | "Sunday":
            return "Weekend"
        
        case _:
            return "Invalid day"
```

```python
day_type("Monday")
```
