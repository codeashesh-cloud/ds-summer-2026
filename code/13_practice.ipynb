### 1. Given this list:
prices = [120, 85, 200, 45, 310, 78, 150]

### Write a for loop that prints each price.
### Then write a for loop that prints ONLY the prices above 100.

```python
prices = [120, 85, 200, 45, 310, 78, 150]
for i in prices:
    print(i)
```

```python
for i in prices:
    if (i>100):
        print(i)
```

```python
# 7. Given:
nums = [4, 7, 2, 9, 1, 8, 3]

# Find the LARGEST number using a loop. 
# Don't use the built-in max() function — do it manually.
# (Hint: assume the first element is the largest, then check each one against it)
```

```python
max=nums[0]
for i in nums:
    if i>max:
        max=i
print(max)
```
