# SETS
a set is an unordered, mutable and unique collection of elements. 

it doesnot allow duplicate values


## creating a set


```python
# empty ste 
a= set()  # we must write set()
print(type(a))

# if we only write the () then it becomes tuple instead of a set
b=()
print(type(b))
```

```python
# set with the elements
c={ 1, 2, 3, 4}
print(c)
print(type(c))
```

```python
#set with different data types
d={1, 2.8, "ashesh", True}
print(d)
print(type(d))
```

```python
#### notes- in above d true is not printed because in python the interpreter assumes true as 1 and we already have 1 in our set.
#so since set doesnot allow duplicate values it doesnot get printed
```

```python
## typecasting or creating the set from a list
numbers= set([1,2,3,2,2,2,45,55,4,4,4,5,5,])
print(numbers) # only unique numbers is printed as no duplicates in the set
```
