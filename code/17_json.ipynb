# Python json module


## what is Json?
JSON(javaScript object notation) is a lightweight data format used for data exchange between servers and applications. it is widely used in APIs, web application and configurations

Python provides the json module to wrok with the json data. 

```python
import json
```

### json.dumps= convert python object (dict) into json string

```python
a={"name": "ALice", "company": "IBM", "salary": 450000}
s=json.dumps(a)
```

```python
print (type(s))
```

```python
print(s)
```

### whereas json.loads= convert the json string into python object (dictionary)

```python
d=json.loads('{"name": "ALice", "company": "IBM", "salary": 450000}')
```

```python
print(type(d))
```

```python
print(d)
```

```python
file= open("data.json", "w")
```

```python
json.dump(a, file)
```
