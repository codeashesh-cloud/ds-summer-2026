# loops


## python has two loops.
## for and while loop


## 1. for loop
it is used to iterate over the sequence like lists, tuples, and strings


```python
a=[1,2,3,4,5,667,7]
```

```python
for i in a:
    print(i)
```

## using range

```python
for i in range (0,10):
    print(i)
```

```python
for i in range(2,20,3): # start , stop, step
    print(i)
```

## 2. while loop 
runs as long as condition is true

```python
count=0
while count <=5:
    print(count)
    count+=1
```

## 3. Loop Control Statements
break → Exits the loop.
continue → Skips to the next iteration.
pass → Does nothing (used as a placeholder).

```python
for i in range(8):
    pass 
```

```python
a=["mango", "banana", "apple", "strawberry"]
for i in a:
    if (i=="apple"):
        break  
    print(i)
    
```

```python
for i in a:
    if (i=="apple"):
        continue
    print(i)
```
