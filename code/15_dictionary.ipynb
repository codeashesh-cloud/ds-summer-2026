# Dictionary 
it is an  ordered(after python 3.7), mutable and key-value pair collection.

it allows efficient data retrieval and modificaton


## creating a dictionary 

```python
#empty dictionary 
a={}
print(type(a))
b=set()
print(type(b))
```

```python
# dictionary with key-value 
student={
    "name": "ashesh",
    "age": 22,
    "grade": "A"
}
print(student)
print(type(student))
    
```

## access= through the key or value


```python
print(student["name"])
# becareful while using this 
#if there is no any key liek we want eg print(student["caste"]). 
#this will throw an error.
```

```python
# so in order to solve that so that we dont get an error if the key is not present in our dictionary , we use .get method
print(student.get("name"))
print(student.get("caste")) # it doenot throw an error 
```

## updating existing value


```python
student["name"]="Ashesh Bhattarai"
print(student)
```
